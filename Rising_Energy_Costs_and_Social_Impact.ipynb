##Tasks:
“How have rising energy costs disproportionately affected low-income households compared to high-income households in the UK since 2021?”



```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data (approximate illustrative values)
quintiles = ['Q1 (Lowest)', 'Q2', 'Q3', 'Q4', 'Q5 (Highest)']
energy_2021 = [8, 5, 4, 3, 2]   # % of income spent on energy in 2021
energy_2024 = [10, 6, 4.5, 3.2, 2.1]  # % of income spent on energy in 2024

x = np.arange(len(quintiles))
width = 0.35

fig, ax = plt.subplots(figsize=(10,6))
bars1 = ax.bar(x - width/2, energy_2021, width, label='2021', color='#1f77b4')
bars2 = ax.bar(x + width/2, energy_2024, width, label='2024', color='#ff7f0e')

ax.set_ylabel('% of Income Spent on Energy')
ax.set_title('Energy Expenditure by Income Quintile (2021 vs 2024)')
ax.set_xticks(x)
ax.set_xticklabels(quintiles)
ax.legend()

# Annotate bars
for bars in [bars1, bars2]:
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:.1f}%', xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0,3), textcoords="offset points", ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
percentage_increase = []
for i in range(len(quintiles)):
    if energy_2021[i] != 0:
        increase = ((energy_2024[i] - energy_2021[i]) / energy_2021[i]) * 100
        percentage_increase.append(increase)
    else:
        percentage_increase.append(float('inf')) # Handle division by zero if 2021 value is 0

print("Percentage Increase in Energy Expenditure (2021 to 2024):")
for i, quintile in enumerate(quintiles):
    if percentage_increase[i] != float('inf'):
        print(f"{quintile}: {percentage_increase[i]:.2f}%")
    else:
        print(f"{quintile}: Infinite increase (2021 value was 0)")
```
